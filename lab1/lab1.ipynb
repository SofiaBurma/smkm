## Чисельне моделювання процесів, які описуються звичайними диференціальними рівняннями

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Модель експоненційного зростання


```python
k = 0.5      
y0 = 1.0      
h = 0.2   
t_end = 5   

t = np.arange(0, t_end, h)
y = np.zeros_like(t)
y[0] = y0

for i in range(len(t) - 1):
    y[i+1] = y[i] + h * k * y[i]
    
t_exact = np.linspace(0, t_end, 200)
y_exact = y0 * np.exp(k * t_exact)
```

```python
plt.plot(t_exact, y_exact, label="Точний розв'язок", linewidth=2)
plt.plot(t, y, label="Метод Ейлера", linewidth=2)
plt.xlabel("t")
plt.ylabel("y(t)")
plt.legend()
plt.grid()
```

### Закон охолодження Ньютона


```python
k = 0.1
temp = 20.0
T0 = 90.0
h = 0.5

t = np.arange(0, 50 + h, h)
T = np.zeros_like(t)
T[0] = T0

for i in range(len(t) - 1):
    T[i+1] = T[i] + h * (-k * (T[i] - temp))

t_exact = np.linspace(0, 50, 200)
T_exact = temp + (T0 - temp) * np.exp(-k * t_exact)
```

```python
plt.plot(t_exact, T_exact, label="Точний розв'язок", linewidth=2)
plt.plot(t, T, label="Метод Ейлера", linewidth=2)
plt.xlabel("t")
plt.ylabel("T(t)")
plt.legend()
plt.grid()
```

### Атрактор Лоренца


```python
sigma = 10.0
r = 28.0
beta = 8.0 / 3.0

h = 0.01
t = np.arange(0, 40, h)

x = np.zeros_like(t)
y = np.zeros_like(t)
z = np.zeros_like(t)

x[0], y[0], z[0] = 1.0, 1.0, 1.0 

for i in range(len(t) - 1):
    dx = sigma * (y[i] - x[i])
    dy = x[i] * (r - z[i]) - y[i]
    dz = x[i] * y[i] - beta * z[i]

    x[i+1] = x[i] + h * dx
    y[i+1] = y[i] + h * dy
    z[i+1] = z[i] + h * dz
```

```python
plt.plot(x, z, lw=0.5)
plt.xlabel("x")
plt.ylabel("z")
plt.grid()
```
